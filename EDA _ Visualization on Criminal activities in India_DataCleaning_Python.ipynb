```python
#Outliers present in the DataFrame: To find Q3 and Q1, we use Pandas quantile method and pass 0.75 and 0.25 values which are 75th and 25th percentile.
def find_outliers_IQR(df):
    
    """This Method Finds the Outliers present in a DataFrame"""

    q1=df.quantile(0.25)

    q3=df.quantile(0.75)

    IQR=q3-q1

    outliers = df[((df<(q1-1.5*IQR)) | (df>(q3+1.5*IQR)))]

    return outliers
```

```python
#outliers for coulmns in Crime_against_women

outliers = find_outliers_IQR(Crime_against_women['2019'])
a = outliers.value_counts().sum()
outliers1 = find_outliers_IQR(Crime_against_women['2020'])
b = outliers1.value_counts().sum()
outliers2 = find_outliers_IQR(Crime_against_women['2021'])
c = outliers2.value_counts().sum()
outliers3 = find_outliers_IQR(Crime_against_women['midPopulation'])
d = outliers3.value_counts().sum()
outliers4 = find_outliers_IQR(Crime_against_women['RateofCrimeAgainstWomen'])
e = outliers4.value_counts().sum()
outliers5 = find_outliers_IQR(Crime_against_women['ChargesSheetingRate'])
f = outliers4.value_counts().sum()
```

```python
outliers_in_Crime_against_women = pd.DataFrame([a,b,c,d,e,f],['2019','2020','2021','midPopulation','RateofCrimeAgainstWomen','ChargesSheetingRate'],columns = ['num of outliers'])
```

```python
outliers_in_Crime_against_women
```

### Exploratory Data Analysis on Datasets Crime_against_women and Crime_against_children

```python
#Which states have the highest number of crimes against women in 2019?

# Group by Name and calculate the sum of crimes in 2019
crimes_women_2019 = Crime_against_women.groupby('Name')['2019'].sum().sort_values(ascending=False)

# Plot the top 10 states with the highest number of crimes against women in 2019
plt.figure(figsize=(10, 6))
sns.barplot(x=crimes_women_2019[:10], y=crimes_women_2019[:10].index)
plt.title("Top 10 States with Highest Crimes Against Women in 2019")
plt.xlabel("Number of Crimes")
plt.ylabel("Name")
plt.show()

crimes_women_2019
```

```python
#What is the overall trend in crimes against women across the years?

# Calculate the total number of crimes against women for each year
total_crimes_women = Crime_against_women[['2019', '2020', '2021']].sum()

# Plot the line chart to show the overall trend in crimes against women over the years
plt.figure(figsize=(10, 6))
sns.lineplot(data=total_crimes_women)
plt.title("Total Crimes Against Women (2019-2021)")
plt.xlabel("Year")
plt.ylabel("Number of Crimes")
plt.show()

total_crimes_women
```

```python
# How does the rate of crime against women compare to the rate of crime against children in 2020?

# Calculate the average rate of crime against women and children in 2020
crime_women_rate_2020 = Crime_against_women['2020'].mean()
crime_children_rate_2020 = Crime_against_children['2020'].mean()

# Create a pie chart to compare the rates of crime against women and children in 2020
plt.figure(figsize=(8, 6))
labels = ['Crime against Women', 'Crime against Children']
sizes = [crime_women_rate_2020, crime_children_rate_2020]
plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title("Comparison of Rates of Crime Against Women and Children in 2020")
plt.show()
```

```python
#What is the overall trend in crimes against women across the years?

# Calculate the total number of crimes against women for each year
total_crimes_children = Crime_against_children[['2019', '2020', '2021']].sum()

# Plot the line chart to show the overall trend in crimes against women over the years
plt.figure(figsize=(10, 6))
sns.lineplot(data=total_crimes_children)
plt.title("Total Crimes Against children (2019-2021)")
plt.xlabel("Year")
plt.ylabel("Number of Crimes")
plt.show()

total_crimes_children
```

```python
# What is the year-wise crimes for the top 10 states against women and children

# Group the data by state and calculate the sum of crimes against women
women_crimes_statewise = Crime_against_women.groupby("Name")[["2019", "2020", "2021"]].sum()

# Group the data by state and calculate the sum of crimes against children
children_crimes_statewise = Crime_against_children.groupby("Name")[["2019", "2020", "2021"]].sum()

top_10_women_states = women_crimes_statewise.sum(axis=1).nlargest(10).index

top_10_children_states = children_crimes_statewise.sum(axis=1).nlargest(10).index

#Filter the data for the top 10 states:
women_crimes_top_10 = women_crimes_statewise.loc[top_10_women_states]
children_crimes_top_10 = children_crimes_statewise.loc[top_10_children_states]

sns.lineplot(data=women_crimes_top_10.T)
plt.xlabel("Year")
plt.ylabel("Number of Crimes")
plt.title("Year-wise Crimes against Women (Top 10 States)")
plt.legend(title="State/UT", loc="upper left", bbox_to_anchor=(1, 1))
plt.show()

sns.lineplot(data=children_crimes_top_10.T)
plt.xlabel("Year")
plt.ylabel("Number of Crimes")
plt.title("Year-wise Crimes against Children (Top 10 States)")
plt.legend(title="State/UT", loc="upper left", bbox_to_anchor=(1, 1))
plt.show()

print(children_crimes_top_10.T)
print(children_crimes_top_10.T)
```
